# Flujo de trabajo Git — Colaboración en pareja



## Tú

```bash
git switch main
git pull origin main       # siempre antes de crear una rama nueva
git switch -c prueba       # creas tu rama de trabajo y te mueves a ella
```

```bash
# ... trabajas en tu zona del archivo ...
git add .
git commit -m "descripción clara de lo que has hecho"
git push -u origin prueba  # primera vez; las siguientes solo: git push
```

```bash
# Abres el PR en VSCode
# Puedes moverte a otra rama y seguir trabajando mientras esperas:
git switch -c prueba-extra
# ... sigues trabajando en otra cosa ...
```

---

## Tu compañero

```bash
# Entra al PR en VSCode, lo revisa y lo acepta (Merge Pull Request)
# Borra la rama remota "prueba" desde VSCode (botón Delete Branch)

git switch main
git pull origin main       # recibe tu trabajo mergeado
git switch su-rama
git merge main             # trae tu trabajo a su rama activa
# ... sigue trabajando ...
```

---

## Tú, después de que el compañero haya mergeado

```bash
git switch main
git pull origin main       # confirmas el merge en tu local
git branch -d prueba       # borras tu rama local, ya no hace falta
git switch -c prueba2      # nueva rama para el siguiente bloque de trabajo
# ... sigues trabajando ...
```

---

## Reglas que nunca hay que romper

1. **Siempre `git pull origin main` antes de crear una rama nueva.**
2. **Commits pequeños y frecuentes**, no acumular todo para el final.
3. **Nunca `git push --force` en main.** Puede borrar el trabajo del otro.
4. **Coordinaos antes de tocar las mismas líneas.** Un mensaje previo evita conflictos.
5. **Nunca resolver un conflicto sin leer las dos versiones.**
